# Hands On Lesson 2

## Import Pandas

```python
import pandas as pd
```

## Load Data

```python
airbnb_test_users = pd.read_csv("airbnb_test_users.csv")
airbnb_test_users.head()
```

## What is the average age of those who use each web browser type?

```python
airbnb_test_users.groupby('first_browser')['age'].mean()
```

## What is the total signup_flow for each device?

```python
airbnb_test_users.groupby('first_device_type')['signup_flow'].sum()
```

## Include the country information from the airbnb_sample_submisson file

### Load airbnb_sample_submission

```python
airbnb_sample_submission = pd.read_csv("airbnb_sample_submission.csv")
airbnb_sample_submission.head()
```

### Merge data on id

```python
airbnb1 = pd.merge(airbnb_test_users, airbnb_sample_submission, on='id')
airbnb1.head()
```

## Include additional users from airbnb_users

### Load data

```python
airbnb_users = pd.read_csv("airbnb_users.csv")
airbnb_users.head()
```

### append airbnb_users

```python
airbnb2 = pd.concat([airbnb1, airbnb_users])
```
